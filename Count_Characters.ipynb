```python
def factorial(number: int) -> int:
    """args:
            Take a single integer number as input
    Returns:
            returns factorial of the number using recursion
    """
    if not isinstance(number, int):
        raise TypeError("Input integer hona chahiye")
    if number < 0:
        raise ValueError("Negative number ka factorial nahi hota")
    if number == 0 or number == 1:
        return 1
    # Recursive case
    return number * factorial(number - 1)

def main():
    try:
        user_input = input("Enter an integer number: ")
        if user_input.strip() == "":
            raise ValueError("No number entered")
        num = int(user_input)
        output = factorial(num)
        print(f"Factorial of {num} is: {output}")

    except (TypeError, ValueError) as e:
        print("Error:", e)
    except Exception as e:
        print("Unexpected Error:", e)
if __name__ == "__main__":
    main()
```

```python
def count_characters(text: str) -> dict:
    """
    args:
        Take a string as input
    Returns:
        returns a dictionary with count of each character
    """
    if not isinstance(text, str):
        raise TypeError("Input string honi chahiye")
    if text.strip() == "":
        raise ValueError("Empty string ka count nahi hota")
    char_count = {}
    for char in text:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1
    return char_count
def main():
    try:
        user_input = input("Enter a string: ")
       # if user_input.strip() == "":
          #  raise ValueError("No string entered")
        result = count_characters(user_input)
        print("Character occurrences:", result)

    except (TypeError, ValueError) as e:
        print("Error:", e)
    except Exception as e:
        print("Unexpected Error:", e)
if __name__ == "__main__":
    main()
```
