```python
def second_largest(num: list[int])-> int:
    """
        find the second largest unique number in the list without using sorting 
    parameters:
                nums(list[int]): a list of integer 
    returns:
            int: second largest uniquew value 
    raise:
            TypeErroe: if input is not a list or element are not integers
            valueeeror: if the list has fewer 2 unique  elemenets
    """
    if not isinstance(num, list):
        raise TypeError("input must be a list of elements.")
    if any(not isinstance(current_number, int) for current_number in num):
        raise TypeError("all input must be integer elements")
    if len(num) < 2:
        raise ValueError("list must have at least two integers")
    if num[0] > num[1]:
        largest = num[0]
        second = num[1]
    else:
        largest = num[1]
        second = num[0]
    for current_number in num[2:]:
        if current_number > largest:
            second = largest
            largest = current_number
        elif current_number != largest and current_number > second:
            second = current_number
    if second == largest:
        raise ValueError("list must contain at least two unique elements")
    return second
if __name__ == "__main__":
    number_list = [1, 1, 4, 1, 5, 1, 1, 10]
    try:
        result = second_largest(number_list)
    except (TypeError, ValueError) as error:
        print(f"error: {error}")
    else:
        print("second largest number:", result)
```
