```python
def remove_duplicates(nums: list[int]) -> list[int]:
    """
    remove duplicate items from a list without using set
    parameters:
        nums (list[int]): a list of integers
    returns:
        list[int]: a list containing only unique elements (original order preserved)
    raise:
        TypeError: if input is not a list or elements are not integers
        ValueError: if the list is empty
    """
    if not isinstance(nums, list):
        raise TypeError("input must be a list")
    if len(nums) == 0:
        raise ValueError("list must not be empty")
    if any(not isinstance(num, int) for num in nums):
        raise TypeError("all elements must be integers")

    unique_list = []
    for num in nums:
        if num not in unique_list:
            unique_list.append(num)
    return unique_list
    
if __name__ == "__main__":
    try:
        size = int(input("enter number of elements: "))
        if size <= 0:
            raise ValueError("size must be greater than zero")

        number_list = []
        for i in range(size):
            value = int(input(f"enter element {i + 1}: "))
            number_list.append(value)

        result = remove_duplicates(number_list)

    except (TypeError, ValueError) as error:
        print(f"error: {error}")
    else:
        print("list after removing duplicates:", result)
```
