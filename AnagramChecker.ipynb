```python
class AnagramChecker:
    """
    This class checks whether two strings are anagrams.
    """
    def __init__(self, str1, str2):
        self.str1 = str1
        self.str2 = str2
    def is_anagram(self):
        """
            args :
                  take a input must be sting 
            retuns:
                    retun the true or flase in check a string is anagram 
        """
        if not isinstance(self.str1, str) or not isinstance(self.str2, str):
            raise TypeError("Input must be strings")

        s1 = self.str1.replace(" ", "").lower()
        s2 = self.str2.replace(" ", "").lower()
        if len(s1) != len(s2):
            return False
        else:
            char_count = {}
            for ch in s1:
                if ch in char_count:
                    char_count[ch] += 1
                else:
                    char_count[ch] = 1
            for ch in s2:
                if ch in char_count and char_count[ch] > 0:
                    char_count[ch] -= 1
                else:
                    return False
            return True
if __name__ == "__main__":
    try:
        string1 = input("Enter first string: ")
        string2 = input("Enter second string: ")
        obj = AnagramChecker(string1, string2)
        if obj.is_anagram():
            print("Strings are ANAGRAMS")
        else:
            print("Strings are NOT anagrams")
    except TypeError as te:
        print("Type Error:", te)
    except Exception as e:
        print("Unexpected Error:", e)
```
