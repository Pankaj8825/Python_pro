```python
def factorial(number: int) -> int:
    """
    args:
        Take a single integer number as input
    Returns:
        returns factorial of the number
    """
    if not isinstance(number, int):
        raise TypeError("Input integer hona chahiye")

    if number < 0:
        raise ValueError("Negative number ka factorial nahi hota")

    result = 1
    for i in range(1, number + 1):
        result = result * i   

    return result
def main():
    try:
        user_input = input("Enter an integer number: ")

        if user_input.strip() == "":
            raise ValueError("No number entered")
        num = int(user_input)
        output = factorial(num)
        print(f"Factorial of {num} is: {output}")

    except (TypeError, ValueError) as e:
        print("Error:", e)
    except Exception as e:
        print("Unexpected Error:", e)

if __name__ == "__main__":
    main()
```
