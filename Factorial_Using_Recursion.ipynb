```python
class Factorial:
    """
    This class calculates factorial of a number using recursion
    """

    def __init__(self, number: int):
        self.number = number
    def calculate(self) -> int:
        """
        args : 
                take a input must be interger
        Returns:
            returns factorial of the number using recursion
        """
        if not isinstance(self.number, int):
            raise TypeError("Input integer hona chahiye")

        if self.number < 0:
            raise ValueError("Negative number ka factorial nahi hota")

        return self._factorial(self.number)

    def _factorial(self, n: int) -> int:
        if n == 0 or n == 1:
            return 1
        return n * self._factorial(n - 1)
def main():
    try:
        user_input = input("Enter an integer number: ")
        if user_input.strip() == "":
            raise ValueError("No number entered")
        num = int(user_input)
        obj = Factorial(num)
        output = obj.calculate()
        print(f"Factorial of {num} is: {output}")
    except (TypeError, ValueError) as e:
        print(e)

if __name__ == "__main__":
    main()
```
