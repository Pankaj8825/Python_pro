```python
class Fibonacci:
    """
    This class prints the first n Fibonacci numbers using iteration.
    """
    def __init__(self, n):
        self.n = n
    def generate(self):
        """
        args: 
            take input in sigle integer numar 
            print the integer number ke fibonicci serise 
        """
        if not isinstance(self.n, int):
            raise TypeError("Input must be an integer")

        if self.n <= 0:
            raise ValueError("n must be a positive integer")
        a, b = 0, 1
        if self.n == 1:
            print(a)
        else:
            for i in range(self.n):
                if i == self.n - 1:
                    print(a, end="")
                else:
                    print(a, end=", ")
                a, b = b, a + b
                
if __name__ == "__main__":
    try:
        n = int(input("Enter value of n: "))
        obj = Fibonacci(n)
        obj.generate()

    except (TypeError,ValueError) as ve:
        print("Value Error:", ve)
    except Exception as e:
        print("Unexpected Error:", e)
```
