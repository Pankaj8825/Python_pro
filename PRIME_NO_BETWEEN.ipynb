```python
def is_prime(number: int) -> bool:
    """
    args:
        Take a single integer number as input
    Returns:
        returns True if number is prime, else False
    """
    if not isinstance(number, int):
        raise TypeError("Input integer hona chahiye")
    if number <= 1:
        return False
    for i in range(2, number):
        if number % i == 0:
            return False
    return True

def main():
    try:
        start = input("Enter starting number: ")
        end = input("Enter ending number: ")

        if start.strip() == "" or end.strip() == "":
            raise ValueError("Range empty nahi honi chahiye")
        start = int(start)
        end = int(end)

        if start > end:
            raise ValueError("Starting number end se chota hona chahiye")

        print(f"Prime numbers between {start} and {end}:")
        for num in range(start, end + 1):
            if is_prime(num):
                print(num, end=" ")
    except (TypeError, ValueError) as e:
        print("Error:", e)
    except Exception as e:
        print("Unexpected Error:", e)

if __name__ == "__main__":
    main()
```
