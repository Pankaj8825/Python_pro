```python
def is_prime(number: int) -> bool:
    """
    args:
        Take a single integer number as input
    Returns:
        returns True if number is prime, else False
    """
    if not isinstance(number, int):
        raise TypeError("Input integer hona chahiye")
    if number <= 1:
        return False
    for i in range(2, number):
        if number % i == 0:
            return False   
    return True
def main():
    try:
        user_input = input("Enter an integer number: ")

        if user_input.strip() == "":
            raise ValueError("No number entered")
        num = int(user_input)
        result = is_prime(num)

        if result:
            print(f"{num} is a Prime Number")
        else:
            print(f"{num} is NOT a Prime Number")

    except (TypeError, ValueError) as e:
        print("Error:", e)
    except Exception as e:
        print("Unexpected Error:", e)

if __name__ == "__main__":
    main()
```
